```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Airbnb_dataset.csv', sep=',')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#df.fillna(0, inplace=True)

df.isnull().sum().sort_values(ascending=False)
```

```python
df['last review'] = df['last review'].fillna('Unknown')
df['reviews per month'] = df['reviews per month'].fillna(0)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['Construction year'] = df['Construction year'].fillna(df['Construction year'].mode()[0])
df['availability 365'] = df['availability 365'].fillna(df['availability 365'].median())
df['review rate number'] = df['review rate number'].fillna(0)
df['minimum nights'] = df['minimum nights'].fillna(df['minimum nights'].median())
```

```python
df['neighbourhood'] = df['neighbourhood'].fillna('Unknown')
df['NAME'] = df['NAME'].fillna('Unknown')
df['number of reviews'] = df['number of reviews'].fillna(0)

# Drop lat/long nulls (important for location)
df = df.dropna(subset=['lat', 'long'])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Remove $ and space, then convert to float
df['price'] = df['price'].str.replace(r'[\$, ]', '', regex=True).astype(float)

# Same for 'service fee' if needed
df['service fee'] = df['service fee'].str.replace(r'[\$, ]', '', regex=True).astype(float)
```

```python
cat_fill = ['host_identity_verified', 'instant_bookable', 'cancellation_policy', 'neighbourhood group', 'host name']
for col in cat_fill:
    df[col] = df[col].fillna('Unknown')

num_fill = ['service fee', 'price', 'calculated host listings count']
for col in num_fill:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['total_cost'] = df['price'] + df['service fee']
```

```python
df.head()
```

```python
df.shape
```

```python
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['last review'] = df['last review'].fillna('Unknown')
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.to_csv("Clean_Airbnb_Data.csv", index=False)
```

```python
cleaned_Airbnb=pd.read_csv('Clean_Airbnb_Data.csv', sep=',')
```

##  🏙️ Heatmap of Correlation Matrix

```python
plt.figure(figsize=(10, 6))
correlation = cleaned_Airbnb[['price','long','service fee', 'number of reviews', 'reviews per month']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## 📈 Average Price by Room Type

```python
plt.flag = plt.figure(figsize=(20,10))
sns.barplot(data=cleaned_Airbnb,x='room type',y='price',estimator='mean',ci=None, palette='Set2')
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()

```

## 🌍 Listings Distribution by Neighbourhood Group

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=cleaned_Airbnb, x='neighbourhood group', order=cleaned_Airbnb['neighbourhood group'].value_counts().index, palette='muted')
plt.title('Listings Count by Neighbourhood Group')
plt.ylabel('Count')
plt.xlabel('Neighbourhood Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 📊 Availability vs. Number of Reviews (Scatter Plot)

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=cleaned_Airbnb, x='availability 365', y='number of reviews', hue='room type', alpha=0.6, palette='cool')
plt.title('Availability vs. Number of Reviews')
plt.xlabel('Availability (365 days)')
plt.ylabel('Number of Reviews')
plt.legend(title='Room Type')
plt.tight_layout()
plt.show()
```

## Average Price by Neighbourhood Group

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=cleaned_Airbnb, x='neighbourhood group', y='price', estimator='mean', palette='magma',errorbar=None)
plt.title('Average Price by Neighbourhood Group')
plt.ylabel('Average Price')
plt.xlabel('Neighbourhood Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

## Room Type Distribution Pie Chart

```python
room_type_counts = cleaned_Airbnb['room type'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Room Type Distribution')
plt.axis('equal')
plt.show()
```
